```python
##Replace schdule SA hours from 8 to 9

for i in range(0,len(df)):
    if df['SA Schedule Hours'][i]==8:
        
        df['SA Schedule Hours'][i] = 9
    else:
        df['SA Schedule Hours'][i]=0
df.head()        
```

```python
##for planned leaves###
                      
##Regulation for planned and unplanned leaves

for i in range(0,len(df)):
    
    ##for planned leaves###
    
## if planned leave= 0, then SA scheduled hours = 9
    
    if df['Is on Planned Leave'][i]== 1 or df['Is on Unplanned leave'][i]== 1:
         df['SA Schedule Hours'][i] = 0  
            
## if planned leave= 0.5, then SA scheduled hours = 4.5  

    elif df['Is on Planned Leave'][i]== 0.5 or df['Is on Unplanned leave'][i]== 0.5:
        df['SA Schedule Hours'][i] = 4.5
        
## if planned leave= 1.0, then SA scheduled hours = 0.0 

    elif df['Is on Planned Leave'][i]== 0 and df['Is on Unplanned leave'][i]== 0:
        df['SA Schedule Hours'][i] = df['SA Schedule Hours'][i]
    else:
        df['SA Schedule Hours'][i] = 0
df.head()        
```

```python
 ##for unplanned leaves ##
for i in range(0,len(df)):    
## if unplanned leave= 0, then SA scheduled hours = 9
    
    if df['Is on Unplanned leave'][i]==0:
         df['SA Schedule Hours'][i] = 9 
            
## if unplanned leave= 0.5, then SA scheduled hours = 4.5  

    elif df['Is on Unplanned leave'][i]==0.5:
        df['SA Schedule Hours'][i] = 4.5
        
## if unplanned leave= 1.0, then SA scheduled hours = 0.0 

    elif df['Is on Unplanned leave'][i]==1:
        df['SA Schedule Hours'][i] = 0   
    
        
    else:
        df['SA Schedule Hours'][i] = df['SA Schedule Hours'][i]
df.head()
#df['SA Schedule Hours'].unique()
```

```python
## iteration for schedule hours



##Replace schdule SA hours from 8 to 9

for i in range(0,len(df)):
    if df['SA Schedule Hours'].values[i]== 8:
        
        df['SA Schedule Hours'].values[i] = 9.0
    else:
        df['SA Schedule Hours'].values[i]=0.0
        
##for planned leaves###
                      
##Regulation for planned and unplanned leaves

for i in range(0,len(df)):
    
    ##for planned leaves###
    
## if planned leave= 0, then SA scheduled hours = 9
    
    if df['Is on Planned Leave'].values[i]==0:
         df['SA Schedule Hours'].values[i] = 9  
            
## if planned leave= 0.5, then SA scheduled hours = 4.5  

    elif df['Is on Planned Leave'].values[i]==0.5:
        df['SA Schedule Hours'].values[i] = 4.5
        
## if planned leave= 1.0, then SA scheduled hours = 0.0 

    elif df['Is on Planned Leave'].values[i]==1:
        df['SA Schedule Hours'].values[i] = 0
        
       ##for unplanned leaves ##
    
## if unplanned leave= 0, then SA scheduled hours = 9
    
    if df['Is on Unplanned leave'].values[i]==0:
         df['SA Schedule Hours'].values[i] = 9 
            
## if unplanned leave= 0.5, then SA scheduled hours = 4.5  

    elif df['Is on Unplanned leave'].values[i]==0.5:
        df['SA Schedule Hours'].values[i] = 4.5
        
## if unplanned leave= 1.0, then SA scheduled hours = 0.0 

    elif df['Is on Unplanned leave'].values[i]==1:
        df['SA Schedule Hours'].values[i] = 0   
    
        
    else:
        df['SA Schedule Hours'].values[i] = 0
df['SA Schedule Hours'].unique()
#df['SA Schedule Hours'].unique()
```

```python
## Replace NA with 0 for logged in hrs

df['Logged in hrs'].unique()
```

```python
##Calculation of updated scheduled hours

for i in range(0,len(df)):
    
    df['Lost Hours'][i]= df['SA Schedule Hours'][i]- df['Logged in hrs'][i]

df.head()
```

```python
##Replace -ve log in hours to 0

for i in range(0,len(df)):
    if df['Lost Hours'][i] < 0:
        df['Lost Hours'][i] = 0
    else:
        df['Lost Hours'][i]=df['Lost Hours'][i]
df.head()       
```

```python
## WoW Site level comparision

sql='''
select week,site,
round(100*(1 - (LostHours)/(SAScheduleHours)),2) as shift_adherence from Schedule_Manager
group by week,site
order by week asc

''';

    
df = pd.read_sql_query(sql,conn)
df.head(20)
```

```python
##Manager level split MAA

sql='''
select week,site,Level1Manager as TM,Level2Manager as OM,round(sum(losthours),2) as losthours,round(sum(SAScheduleHours),2) as schedulehours,
round(100*(1 - sum(LostHours)/sum(SAScheduleHours)),2) as shift_adherence from Schedule_Manager
group by week,level1manager
order by week asc

''';

    
df_sql_result = pd.read_sql_query(sql,conn)
df_sql_result
```
