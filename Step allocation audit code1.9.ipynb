```python
import pandas as pd
import numpy  as np

##import the input files

df_aj= pd.read_excel(r'C:\Users\arvinmvn\Desktop\Dhruv automation task\aj_input.xlsx',)
df_pardot= pd.read_excel(r'C:\Users\arvinmvn\Desktop\Dhruv automation task\pardot_input.xlsx',)


df_pardot.head()
# df_aj.head()
```

```python
## create a req list

## add the A+ service name in the filter for all the eligble use cases (Select one of the services for Aplus should not be I do not want A+ cataloging)


df_pardot_a_plus=df_pardot.copy() ## filter A+ catloging service

df_pardot_a_plus['Select one of the SPN services']='Enhanced Brand Content' ## add enchanced brand content as service

all_df_list=[df_pardot_a_plus,df_pardot]  ## add the a+ list to existing pardot  file

df_pardot_updated = pd.concat(all_df_list)

df_req_list=df_pardot_updated.sort_values(by=['STEP_SPN_Merchant'])
df_req_list.head()
```

```python
## filter the aj file, take only requestsource as 'Referral' and requestsourcereftag as 'Step'

df_aj=df_aj[df_aj['requestsource']=='Referral']

df_aj=df_aj[df_aj['requestsourcereftag']=='STEP Program']


## rename Advertisement Optimization and Cataloging to match the service names for req_list df

df_aj['Service Name']=df_aj['Service Name'].replace('Advertising Optimization','Advertisement Optimization')
df_aj['Service Name']=df_aj['Service Name'].replace('Cataloguing','Cataloging')


df_aj.head()
```

```python
## create the pseudo primary key of step merchant id + servicename


df_aj['sss']=df_aj['sellercustomerid'].astype(str)+ df_aj['Service Name'].astype(str) 

df_req_list['sss']=df_req_list['STEP_SPN_Merchant'].astype(str)+df_req_list['Select one of the SPN services'].astype(str)
df_req_list.head()

# df_req_list['key_two'].head()
```

```python
## join the created keys and extract the parts 

df_result=pd.merge(df_req_list, df_aj, on="sss",how='left')

df_result.head()
```

```python
df_result=df_result[['Last','STEP_SPN_Merchant','STEP_SPN_Name','Email','STEP_SPN_Contact','Select one of the SPN services','SP Name','sss','sellercustomerid']]

df_result.head()
```

```python
df_result['Created']=''
df_result['sellercustomerid']=df_result['sellercustomerid'].replace(np.nan, 'empty')
for i in range (0,len(df_result)):
    if df_result['sellercustomerid'][i]=='empty':
        df_result['Created'][i]='No'
    else:
        df_result['Created'][i]='Yes'
df_result.head()        
```
