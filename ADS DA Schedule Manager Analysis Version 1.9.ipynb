```python
## Add week, Schedule login timestamp and first login timestamp

df.insert(1, "Week","")
df.insert(2, "Schedule Login TimeStamp","")
df.insert(3, "First login time TimeStamp","")
df.insert(4, "Login Deficit","")
df.insert(5, "Late login or Not","")
df.head()
```

```python
##iteration for date to week conversion

df['Calendar Date']=pd.to_datetime(df['Calendar Date'],format='%Y-%m-%d %H:%M:%S')
df['Calendar Date']


## Map corresponding week from date

for i in range(0,len(df)):
     df['Week'][i] = df['Calendar Date'][i].week
df.head()
```

```python
## Week confirmation

df['Week'].unique()
```

```python
##Schedule Login datetime to timestamp format

df['Schedule Login Time']=pd.to_datetime(df['Schedule Login Time'],format='%Y-%m-%d %H:%M:%S')

for i in range(0,len(df)):
    df['Schedule Login TimeStamp'][i] = df['Schedule Login Time'][i].strftime("%H:%M:%S")
df.head()
```

```python
##First Login contains null values, to handle the missing data we shall use the value of the login time
## the login deficit calculation for this will be 0, therefore it will not impede our result calculation
# for the count of late logins
df['First login time'] = df['First login time'].fillna(df['Schedule Login Time'])
```

```python
##First Login datetime to timestamp format

df['First login time']=pd.to_datetime(df['First login time'],format='%Y-%m-%d %H:%M:%S')

for i in range(0,len(df)):
    df['First login time TimeStamp'][i] = df['First login time'][i].strftime("%H:%M:%S")
df.head()
```

```python
##Login Deficit calculation

for i in range(len(df)):
            sh=  str(df['Schedule Login TimeStamp'].values[i])
            sv=  str(df['First login time TimeStamp'].values[i])
            h,m,s = sh.split(':' or '.')
            p,r,b = sv.split(':' or '.')
            df['Login Deficit'].values[i]= float(dt.timedelta(hours=int(p),minutes=int(r),seconds=float(b)).total_seconds()/60)- float(dt.timedelta(hours=int(h),minutes=int(m),seconds=float(s)).total_seconds()/60)


df.head()        
```

```python
##Late login or not

for i in range(0,len(df)):
    if df['Login Deficit'][i] > 3:
        df['Late login or Not'][i]='Late Login'
    else:
        df['Late login or Not'][i]='Not Late Login'
df.head()       
```

```python
##calculate updated lost hours

## insert Schedule_hours_updated column

df.insert(39, "Lost Hours","")
df.head()
```

```python
##for planned leaves###
                      
##Regulation for planned and unplanned leaves

for i in range(0,len(df)):
    
    ##for planned leaves###
    
## if planned leave= 0, then SA scheduled hours = 8.0
    
    if df['Is on Planned Leave'][i]== 1 or df['Is on Unplanned leave'][i]== 1:
         df['SA Schedule Hours'][i] = 0  
            
## if planned leave= 0.5, then SA scheduled hours = 4.0  

    elif df['Is on Planned Leave'][i]== 0.5 or df['Is on Unplanned leave'][i]== 0.5:
        df['SA Schedule Hours'][i] = 4.0
        
## if planned leave= 1.0, then SA scheduled hours = 0.0 

    elif df['Is on Planned Leave'][i]== 0 and df['Is on Unplanned leave'][i]== 0:
        df['SA Schedule Hours'][i] = df['SA Schedule Hours'][i]
    else:
        df['SA Schedule Hours'][i] = 0
df.head()        
```
