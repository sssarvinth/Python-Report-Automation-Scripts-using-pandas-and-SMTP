```python
##import liberaries
import pandas as pd
import numpy as np
import datetime as dt
df= pd.read_excel(r'C:\Users\arvinmvn\Desktop\AWS SAMMI Reports\AWS Verification Velocity Report\Velocity Automation.xlsx',)
df.head()
```

```python
#insert processed hours column
df.insert(9, "File compare","")
df.insert(10, "Processed Hours","")
df.head()
```

```python
## compare Audio length and previous audio length. for example 2nd row is compared to 1st row, if match true is returned in the second column

df['File compare'] = df['Audio Duration\r\n(HH:MM:SS.mss)'].eq(df['Audio Duration\r\n(HH:MM:SS.mss)'].shift())

df.head()
```

```python
## change the true or false to 1 or 0, match is true, mismatch is false
for column in df['File compare']:
    df['File compare'] = df['File compare'].replace(True,1)
    df['File compare'] = df['File compare'].replace(False,0) 
df.head() 


 
```

```python
##Working code ..SSS!!!!, the processed audio hours is show in minutes

## if validation = false, then they are two different files so File length = processed length, else processed length = audio length - completed length
#df.head() 
for i in range(len(df)):
        if df['File compare'][i] == 1.0:    
            #df['Processed Hours'][i]= pd.to_datetime(df['Audio Duration\r\n(HH:MM:SS.mss)'][i]) - pd.to_datetime(df['Completed Duration (HH:MM:SS.mss)'][i])
            # df['Column'].values[i] 
           # df['Processed Hours'][i]= df['Number Of Speakers'][i]
            sh = str(df['Completed Duration (HH:MM:SS.mss)'].values[i])
            sv=  str(df['Completed Duration (HH:MM:SS.mss)'].values[i-1])
            h,m,s = sh.split(':' or '.')
            p,r,b = sv.split(':' or '.')
            df['Processed Hours'][i]= float(dt.timedelta(hours=int(h),minutes=int(m),seconds=float(s)).total_seconds()/60)- float(dt.timedelta(hours=int(p),minutes=int(r),seconds=float(b)).total_seconds()/60)
            i=i+1
            # print('hi')                       
           
        else:
            iv = str(df['Completed Duration (HH:MM:SS.mss)'].values[i])
            a,k,y = iv.split(':' or '.')
            
            df['Processed Hours'][i]= float(dt.timedelta(hours=int(a),minutes=int(k),seconds=float(y)).total_seconds()/60)
            i=i+1
             #print('hie i am 1')


df.head()
```
