```python
## Import Liberaries
import pandas as pd
import numpy as np
df= pd.read_excel(r'C:\Users\arvinmvn\Desktop\AWS SAMMI Reports\Dashboard\AWS Verification Scorecard-RAW June.xlsx')
df.head()
```

```python
##Rename the columns
df = df.rename(columns = {'Aht=  ( Autolog in millieseconds)/(Processed hours in millieseconds)': 'AHT','Encryption Score': 'ES'}, inplace = False)
df.head()
```

```python
##Change overturn % to percent

for i in range(0,len(df)):
    df['Overturn %'].values[i]=100*(df['Overturn %'].values[i])
df.head()    
```

```python
## Pivot Data
table = pd.pivot_table(df[df['Week']==25], values=['Processed Hours','AHT','ES','Overturn %'], index=['User ID'],margins=False,
               margins_name='total', aggfunc=sum , fill_value=0)
table.round(2)
```

```python
## bulk mail for hourly processed coint (hourly velocity)

import win32com.client as win32
outlook = win32.Dispatch('outlook.application')
for i in range(0,len(table)):
    mail = outlook.CreateItem(0)

    mail.Subject = 'AWS Verification Performance Summary Report'
    
    
    mail.To = str(table.index[i])+'@amazon.com' 
    mail.CC = 'dmaniva@amazon.com'
    
## i-> position of data in dataframe
    mail.HTMLBody = '''<h3>Hi, Please find your AWS Verification Performance Summary for the week.</h3>
                   {}'''.format(table.round(2)[i:i+1].to_html())  ## : i + 1 increment 1 of index 
# To attach a file to the email (optional):
#attachment  = "Path to the attachment"
#mail.Attachments.Add(attachment)

    mail.Send()
```
