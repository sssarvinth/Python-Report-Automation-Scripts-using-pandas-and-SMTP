```python
##1. Shift hours Report

import pandas as pd

df= pd.read_excel(r'C:\Users\arvinmvn\Desktop\ADS Login Managment System.xlsx')
df.head()
```

```python
##filter the weeks
df = df[df['Week'].between(13,16)]

##view pivot
table = pd.pivot_table(df, values=['Shift Hours'], index=['User Alias'], columns= ['Date'] ,margins=False,
               margins_name='total', aggfunc=sum , fill_value=0)
## display all options
#pd.set_option('display.max_rows', df.shape[0]+1)
table.round(2)
```

```python
##for bulk email 
import win32com.client as win32
outlook = win32.Dispatch('outlook.application')
for i in range(0,len(table)):
    mail = outlook.CreateItem(0)

    mail.Subject = 'AWS Velocity Verifiction Report'


    mail.To = str(table.index[i])+'@amazon.com' 
    mail.CC = 'thenmozs@amazon.com'
    
## i-> position of data in dataframe
    mail.HTMLBody = '''<h3>Hi, Please find your processed hours for the week.</h3>
                   {}'''.format(table[i:i+1].to_html())  ## : i + 1 increment 1 of index 
# To attach a file to the email (optional):
#attachment  = "Path to the attachment"
#mail.Attachments.Add(attachment)

    mail.Send()
```
