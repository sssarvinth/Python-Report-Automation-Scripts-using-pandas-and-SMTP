```python
import pandas as pd
import pandas as pd
import numpy as np
import datetime as dt


df1= pd.read_csv(r'C:\Users\arvinmvn\AppData\Local\Temp\sss.csv',)
```

```python
##drop the null values

df1.dropna(subset = ["Entry Date"], inplace=True)
df1.head()
```

```python
table = pd.pivot_table(df1, values=['Production (in hrs)'], index=['User ID'], columns= ['Date'] ,margins=False,
               margins_name='total', aggfunc=sum , fill_value=0)
table.round(2)
```

```python
for i in range(0,len(df1)):
            sh=  str(df1['End Time'].values[i])
            sv=  str(df1['Start Time'].values[i])
            h,m,s = sh.split(':' or '.')
            p,r,b = sv.split(':' or '.')
            df1['Autolog Hours'].values[i]= float(dt.timedelta(hours=int(h),minutes=int(m),seconds=float(s)).total_seconds()/3600)- float(dt.timedelta(hours=int(p),minutes=int(r),seconds=float(b)).total_seconds()/3600)
            i=i+1
```
